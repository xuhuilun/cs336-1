```python
str = '你好'

byte_str = str.encode('utf-8')
print(f"原始字符串: {str}")
print(f"UTF-8 编码的字节串: {byte_str}")
print(f"字节列表: {[b for b in byte_str]}")
```

```python
a = '你'.encode('utf-8')
b = '好'.encode('utf-8')
print(f"'你' 的 UTF-8 编码: {a}, 字节列表: {[byte for byte in a]}")
print(f"'好' 的 UTF-8 编码: {b}, 字节列表: {[byte for byte in b]}")

print(f"合并后的字节串: {a + b}")
print(f"合并后的字节串的字节列表: {[byte for byte in a + b]}")
```

```python
import regex as re

gpt2_pat = re.compile(r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+""")
text = "您好 人没了"
text = "Hello World test! ..."
tokens = gpt2_pat.findall(text)
print(tokens)
```

```python
special_tokens = ["<special1>", "<special2>", "<special3>"]

text = "这是一个测试文本，包含<special1>一些特殊的<special2>标记<special3>。"

special_regex = "|".join(re.escape(t) for t in special_tokens)
# 使用 re.split 进行分割。关键是使用捕获组 `(...)`，这样特殊 Token 本身也会被保留在结果列表中。
parts = re.split(f"({special_regex})", text)
train_segments = [p for p in parts if p not in special_tokens]
print(parts)
print(train_segments)
```

```python
bs = list(range(ord("!"), ord("~") + 1)) + list(range(ord("¡"), ord("¬") + 1)) + list(range(ord("®"), ord("ÿ") + 1))
cs = bs[:]
n = 0
for b in range(256):
    if b not in bs:
        bs.append(b)
        cs.append(256 + n)
        n += 1
cs = [chr(n) for n in cs]

byte_to_unicode = dict(zip(bs, cs))
print(byte_to_unicode)
```
