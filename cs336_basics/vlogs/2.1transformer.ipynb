```python
import time
import torch

model = torch.nn.Linear(1024, 1024).cuda()
# 模拟单句
x_single = torch.randn(1, 1024).cuda()
# 模拟批处理
x_batch = torch.randn(64, 1024).cuda()

# 测试单句耗时
torch.cuda.synchronize()
start = time.time()
for _ in range(100):
    _ = model(x_single)
torch.cuda.synchronize()
print(f"单句 100 次耗时: {time.time() - start:.4f}s")

# 测试批处理耗时
start = time.time()
for _ in range(100):
    _ = model(x_batch)
torch.cuda.synchronize()
print(f"64 句 100 次耗时: {time.time() - start:.4f}s")
```
