```python
char_en = "A"
print(f"字符: {char_en}, 内存(ASCII/Unicode): {ord(char_en)}")

char_cn = "牛"
print(f"字符: {char_cn}, Unicode码点(十进制): {ord(char_cn)}")
print(f"字符: {char_cn}, Unicode码点(十六进制): {hex(ord(char_cn))}")
```

```python
text = "牛"

# 编码: 字符 -> 字节
encoded_bytes = text.encode("utf-8")
print(f"原始文本: {text}")
print(f"UTF-8 字节流: {encoded_bytes}")
print(f"字节列表: {[b for b in encoded_bytes]}")

# 'A' (65) -> 0x41 (十进制65) -> 1个字节
# '牛' -> 0xe7, 0x89, 0x9b -> 3个字节
```

```python
print(ord('牛')) # 29275
print(hex(29275)) # 0x725B
print(bin(29275)) # '0b111001001011011'
print('牛'.encode('utf-8')) # b'\xe7\x89\x9b'
```
