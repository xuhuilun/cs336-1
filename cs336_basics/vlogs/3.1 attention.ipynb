```python
import torch

# 模拟两组不同的得分
z_small = torch.tensor([2.0, 1.0, 0.0], requires_grad=True)
z_large = torch.tensor([50.0, 20.0, 0.0], requires_grad=True)

# 计算 Softmax
p_small = torch.softmax(z_small, dim=0)
p_large = torch.softmax(z_large, dim=0)

# 对最大值尝试进行反向传播
p_small[0].backward()
grad_small = z_small.grad.clone()

z_large.grad = None # 清空梯度
p_large[0].backward()
grad_large = z_large.grad

print(f"小数值下的梯度: {grad_small}") 
# 输出: [ 0.22, -0.16, -0.06] -> 梯度很活跃

print(f"大数值(One-hot)下的梯度: {grad_large}") 
# 输出: [ 0.00,  0.00,  0.00] -> 梯度死了！
```
